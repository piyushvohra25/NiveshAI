```python
import pandas as pd
import os

all_data = []

for file in files:

    df = pd.read_csv(
        os.path.join(folder_path, file)
    )

    company = file.replace(".csv", "")

    df["Company"] = company

    all_data.append(df)

master_df = pd.concat(
    all_data,
    ignore_index=True
)

print("Shape:", master_df.shape)

master_df.head()
```

```python
model_df = master_df.copy()

model_df = model_df.dropna(
    subset=[
        "Momentum_5",
        "MA20",
        "MA50",
        "MA252",
        "MA20_Ratio",
        "MA50_Ratio",
        "MA252_Ratio",
        "Future_Return_63",
        "Future_Return_126",
        "Future_Return_252",
        "Future_Return_630",
        "Future_Return_1260"
    ]
)
```

```python
model_df = model_df.sort_values(["Company", "Date"]).reset_index(drop=True)
```

```python
features = [
    "Prev Close",
    "Open",
    "High",
    "Low",
    "Close",
    "VWAP",
    "Volume",
    "Turnover",
    "Daily_Return",
    "Volatility",
    "Momentum_5",
    "MA20_Ratio",
    "MA50_Ratio",
    "MA252_Ratio"
]

target_cols = [
    "Future_Return_63",
    "Future_Return_126",
    "Future_Return_252",
    "Future_Return_630",
    "Future_Return_1260"
]
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import joblib
```

```python
BASE_DIR = "/content"   # or your working dir

os.makedirs(f"{BASE_DIR}/models", exist_ok=True)
os.makedirs(f"{BASE_DIR}/scalers", exist_ok=True)
```

```python
latest_features = (
    model_df
    .sort_values(["Company", "Date"])
    .groupby("Company")
    .tail(1)
)

latest_features.to_csv(
    "latest_features.csv",
    index=False
)

print(latest_features.shape)
```

```python
import tensorflow as tf
import keras
```

```python
for company in model_df["Company"].unique():

    print(f"\nProcessing {company}")

    df = model_df[model_df["Company"] == company].copy()
    df = df.dropna(subset=features + target_cols)

    if len(df) < 500:
        print(f"Skipping {company} (not enough data)")
        continue

    X = df[features]
    y = df[target_cols]

    split = int(len(df) * 0.8)

    X_train = X.iloc[:split]
    X_test = X.iloc[split:]

    y_train = y.iloc[:split]
    y_test = y.iloc[split:]

    scaler = StandardScaler()

    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(5)
    ])

    model.compile(optimizer='adam', loss='mse', metrics=['mae'])

    model.fit(
    X_train_scaled,
    y_train,
    validation_data=(X_test_scaled, y_test),
    epochs=15,
    batch_size=32
    )

    model.evaluate(X_test_scaled, y_test)

    model.save(f"{BASE_DIR}/models/{company}_model.keras")
    joblib.dump(scaler, f"{BASE_DIR}/scalers/{company}_scaler.pkl")
```
